# Data Normalization Statistics Extraction

This notebook loads the behavioral biometric feature dataset, calculates the mean and standard deviation for each feature, and saves these statistics to a CSV file. These parameters are essential for future Z-score normalization (Standardization) of the data.

```python
import pandas as pd
import numpy as np
import os

INPUT_FILE = r'../../data_processed/feature_kmt_dataset_Edge_Hill_University_22/feature_extraction_V3/feature_extraction_V3_comprehensive_88_users.csv'

if not os.path.exists(INPUT_FILE):
    print(f"Error: {INPUT_FILE} not found.")
else:
    df = pd.read_csv(INPUT_FILE)
    print("EDA summary for:", INPUT_FILE)
    print("Shape:", df.shape)

    print("\nColumn types count:")
    print(df.dtypes.value_counts())

    print("\nNumeric columns count:", len(df.select_dtypes(include=[np.number]).columns))
    print("Non-numeric columns count:", len(df.select_dtypes(exclude=[np.number]).columns))

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nMissing values per column:")
    missing = df.isna().sum()
    print(missing[missing > 0])
```

```python
import pandas as pd
import numpy as np
import os

# 1. Load the feature dataset
# Note: The user requested 'feature_importance_ranking_88_users.csv', 
# but for normalization we typically need the feature values from the extraction file.
# We will load the specified file and calculate stats for its numeric columns.
INPUT_FILE = r'../../data_processed/feature_kmt_dataset_Edge_Hill_University_22/feature_extraction_V3/feature_extraction_V3_comprehensive_88_users.csv'
OUTPUT_FILE = r'../../data_processed/feature_kmt_dataset_Edge_Hill_University_22/mean_stad_for_norm.csv'

if not os.path.exists(INPUT_FILE):
    print(f"Error: {INPUT_FILE} not found.")
else:
    df = pd.read_csv(INPUT_FILE)
    print(f"Loaded dataset with shape: {df.shape}")

    # 2. Select numeric columns (features) and drop identifiers if they are not to be normalized
    # We exclude user_id, trial_id, label, and timestamp from normalization statistics
    exclude_cols = ['user_id', 'trial_id', 'label', 'timestamp']
    numeric_cols = df.select_dtypes(include=[np.number]).columns.difference(exclude_cols)
    numeric_df = df[numeric_cols]

    # 3. Calculate Mean and Standard Deviation for each feature
    stats_df = pd.DataFrame({
        'Feature': numeric_df.columns,
        'Mean': numeric_df.mean(),
        'Std': numeric_df.std()
    }).reset_index(drop=True)

    print("\nNormalization Statistics (First 10 features):")
    print(stats_df.head(10))

    # 4. Save to CSV
    stats_df.to_csv(OUTPUT_FILE, index=False)
    print(f"\nNormalization statistics saved to: {OUTPUT_FILE}")
```
