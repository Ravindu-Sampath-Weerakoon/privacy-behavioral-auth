# Advanced Behavioral Biometric Feature Analysis (V2) - Parametric Fitting

This notebook enhances the feature extraction process by fitting Keystroke and Mouse Trajectory (KMT) latency data to parametric distributions (**Gamma** and **Log-Normal**). 

### Why Parametric Fitting?
Human timing behavior (keystrokes, latencies) is rarely perfectly normal. It typically exhibits a right-skewed distribution. By fitting these to specific distributions, we can extract:
- **Shape (a/k)**: Captures the 'peakedness' and skewness of the user's behavior.
- **Location (loc)**: Represents the minimum offset (baseline reaction time).
- **Scale (scale)**: Captures the spread or variance of the timings.

These parameters are more stable and harder to spoof than simple means or standard deviations.

```python
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import gamma, lognorm

import warnings
warnings.filterwarnings('ignore')

# Configuration
RAW_DATA_PATH = r'../data_raw/feature_kmt_dataset_Edge_Hill_University_22/feature_kmt_json'
USERS_TO_LOAD = 5  # Demonstrating with first 5 users
```

## 1. Data Processing Functions

We maintain privacy by extracting only timing latencies and dropping actual key strings.

```python
def get_timings(events):
    """
    Extracts Dwell Times and Digraph Latencies with IQR Outlier Removal.
    """
    last_press = {}
    raw_dwells = []
    raw_digraphs = []
    last_press_time = None

    for event in events:
        key = event.get('Key')
        event_type = event.get('Event')
        try:
            epoch = float(event.get('Epoch'))
        except (ValueError, TypeError):
            continue
            
        if event_type == 'pressed':
            last_press[key] = epoch
            if last_press_time is not None:
                latency = epoch - last_press_time
                if 0 < latency < 5: # Basic range filter
                    raw_digraphs.append(latency)
            last_press_time = epoch
        elif event_type == 'released':
            if key in last_press:
                dwell = epoch - last_press[key]
                if 0 < dwell < 2: # Basic range filter
                    raw_dwells.append(dwell)
                del last_press[key]
    
    def remove_iqr_outliers(data):
        if len(data) < 10: return np.array(data)
        q1, q3 = np.percentile(data, [25, 75])
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        # In biometrics, we focus on the upper bound (slow distractions)
        return np.array([x for x in data if max(0.01, lower_bound) <= x <= upper_bound])

    return remove_iqr_outliers(raw_dwells), remove_iqr_outliers(raw_digraphs)

def fit_best_distribution(data):
    """
    Fits Gamma and Log-Normal distributions and returns parameters for both.
    Also returns a flag indicating which one has a better log-likelihood.
    """
    if len(data) < 5:
        return [np.nan] * 7

    # Gamma Fit
    g_shape, g_loc, g_scale = gamma.fit(data, floc=0)
    g_loglik = np.sum(gamma.logpdf(data, g_shape, loc=g_loc, scale=g_scale))

    # Log-Normal Fit
    l_s, l_loc, l_scale = lognorm.fit(data, floc=0)
    l_loglik = np.sum(lognorm.logpdf(data, l_s, loc=l_loc, scale=l_scale))

    better_fit = 'Gamma' if g_loglik > l_loglik else 'Log-Normal'
    
    return [g_shape, g_loc, g_scale, l_s, l_loc, l_scale, better_fit]
```

## 2. Feature Extraction Pipeline

We process each trial (session) and calculate parametric features for both Dwell Times and Digraph Latencies. We also store the raw timings for visualization.

```python
all_features = []
raw_timings_store = {} # Stores raw data for KDE plotting
file_list = sorted([f for f in os.listdir(RAW_DATA_PATH) if f.endswith('.json')])[:USERS_TO_LOAD]

for file_name in file_list:
    file_path = os.path.join(RAW_DATA_PATH, file_name)
    with open(file_path, 'r') as f:
        data = json.load(f)
        user_id = file_name.split('_')[-1].replace('.json', '')
        raw_timings_store[user_id] = {'true': {}, 'false': {}}
        
        # Process True and False data
        for category in ['true_data', 'false_data']:
            cat_key = 'true' if category == 'true_data' else 'false'
            label = 1 if category == 'true_data' else 0
            for trial_id, trial_data in data.get(category, {}).items():
                events = trial_data.get('key_events', [])
                dwells, digraphs = get_timings(events)
                
                # Store raw for plot
                raw_timings_store[user_id][cat_key][trial_id] = {'dwells': dwells, 'digraphs': digraphs}
                
                # Fit Dwell Times
                dwell_params = fit_best_distribution(dwells)
                # Fit Digraph Latencies
                digraph_params = fit_best_distribution(digraphs)
                
                feature_row = {
                    'user_id': user_id,
                    'trial_id': trial_id,
                    'label': label,
                    # Dwell Parametric Features
                    'dwell_gamma_shape': dwell_params[0],
                    'dwell_gamma_loc': dwell_params[1],
                    'dwell_gamma_scale': dwell_params[2],
                    'dwell_lognorm_s': dwell_params[3],
                    'dwell_lognorm_loc': dwell_params[4],
                    'dwell_lognorm_scale': dwell_params[5],
                    'dwell_best_fit': dwell_params[6],
                    # Digraph Parametric Features
                    'digraph_gamma_shape': digraph_params[0],
                    'digraph_gamma_loc': digraph_params[1],
                    'digraph_gamma_scale': digraph_params[2],
                    'digraph_lognorm_s': digraph_params[3],
                    'digraph_lognorm_loc': digraph_params[4],
                    'digraph_lognorm_scale': digraph_params[5],
                    'digraph_best_fit': digraph_params[6]
                }
                all_features.append(feature_row)

df_features = pd.DataFrame(all_features)
print(f"Extracted parametric features for {len(df_features)} trials.")
display(df_features.head())
```

## 3. Visualization: Individual Trial Fits

We visualize the raw data distribution (Histogram + KDE) against the theoretical PDF curves for a single trial.

```python
def plot_fitted_distribution(user_id, trial_id, feature_type='dwell'):
    # Retrieve raw data
    trial_data = raw_timings_store.get(user_id, {}).get('true', {}).get(trial_id)
    if not trial_data:
        trial_data = raw_timings_store.get(user_id, {}).get('false', {}).get(trial_id)
    
    if not trial_data: 
        print(f"Trial {trial_id} not found for user {user_id}")
        return
        
    plot_data = trial_data['dwells'] if feature_type == 'dwell' else trial_data['digraphs']
    if len(plot_data) < 5: 
        print("Not enough data to plot")
        return
    
    # Get parameters from our dataframe
    row = df_features[(df_features['user_id'] == user_id) & (df_features['trial_id'] == trial_id)].iloc[0]
    
    plt.figure(figsize=(10, 5))
    
    # 1. Raw Data: Histogram + KDE
    sns.histplot(plot_data, bins=20, stat='density', alpha=0.3, color='gray', label='Raw Data Histogram')
    sns.kdeplot(plot_data, color='black', lw=1.5, label='Raw Data KDE', linestyle=':')
    
    # 2. Theoretical PDF Curves
    x = np.linspace(min(plot_data), max(plot_data), 100)
    
    # Plot Gamma
    y_gamma = stats.gamma.pdf(x, row[f'{feature_type}_gamma_shape'], loc=row[f'{feature_type}_gamma_loc'], scale=row[f'{feature_type}_gamma_scale'])
    plt.plot(x, y_gamma, 'r-', lw=2, label=f'Fitted Gamma (Best: {row[feature_type + "_best_fit"] == "Gamma"})')
    
    # Plot Log-Normal
    y_lognorm = stats.lognorm.pdf(x, row[f'{feature_type}_lognorm_s'], loc=row[f'{feature_type}_lognorm_loc'], scale=row[f'{feature_type}_lognorm_scale'])
    plt.plot(x, y_lognorm, 'b--', lw=2, label=f'Fitted Log-Normal (Best: {row[feature_type + "_best_fit"] == "Log-Normal"})')
    
    plt.title(f'User {user_id} - {trial_id} - {feature_type.capitalize()} Timings Parametric Fit')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Visualize first user, first trial
plot_fitted_distribution('0001', 'test_1', feature_type='dwell')
plot_fitted_distribution('0001', 'test_1', feature_type='digraph')
```

## 4. Multi-User Profile Comparison (Advanced Visualization)

This visualization overlays the theoretical **Gamma PDF** curves of different users. Crucially, it also plots the **Raw Data KDEs** for every trial in a lighter shade of the same color. This demonstrates the consistency of the user's behavior and the robustness of the parametric model.

```python
def plot_multi_user_comparison(user_list, feature_type='dwell'):
    plt.figure(figsize=(12, 8))
    colors = sns.color_palette('husl', len(user_list))
    
    x = np.linspace(0, 0.6, 250) 

    for idx, user_id in enumerate(user_list):
        color = colors[idx]
        
        # 1. Plot individual trial raw distributions (faint lines)
        trials = raw_timings_store.get(user_id, {}).get('true', {})
        for t_id, t_data in trials.items():
            data = t_data['dwells'] if feature_type == 'dwell' else t_data['digraphs']
            if len(data) > 1:
                sns.kdeplot(data, color=color, alpha=0.2, lw=1)
        
        # 2. Plot Theoretical Average Profile (Bold Line)
        user_rows = df_features[(df_features['user_id'] == user_id) & (df_features['label'] == 1)]
        if user_rows.empty: continue
        
        avg_shape = user_rows[f'{feature_type}_gamma_shape'].mean()
        avg_loc = user_rows[f'{feature_type}_gamma_loc'].mean()
        avg_scale = user_rows[f'{feature_type}_gamma_scale'].mean()

        y_pdf = stats.gamma.pdf(x, avg_shape, loc=avg_loc, scale=avg_scale)
        plt.plot(x, y_pdf, color=color, lw=3.5, label=f'User {user_id} Theoretical PDF')
        
        # Optional: Shading
        plt.fill_between(x, y_pdf, alpha=0.05, color=color)

    plt.title(f'Multi-User Biometric Signatures: Raw KDEs vs. Theoretical {feature_type.capitalize()} Profiles', fontsize=16)
    plt.xlabel('Time (seconds)', fontsize=13)
    plt.ylabel('Density', fontsize=13)
    plt.xlim(0, 0.5)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Compare First 5 Users for both Dwell and Digraph
top_5_users = sorted(df_features['user_id'].unique())[:5]
print("Comparing Dwell Time Profiles...")
plot_multi_user_comparison(top_5_users, feature_type='dwell')

print("\nComparing Digraph Latency Profiles...")
plot_multi_user_comparison(top_5_users, feature_type='digraph')
```

## 5. Hierarchical Clustering of User Signatures

In this section, we use the extracted parametric features to see if trials belonging to the same user naturally group together. This is a crucial step for **user identification** and **authentication**.

### Why Hierarchical Clustering?
- It allows us to visualize the 'closeness' of different trials via a **Dendrogram**.
- It doesn't require us to pre-specify the number of clusters (though we expect 5).
- We use **Ward's Linkage** to minimize variance within each cluster.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# 1. Filter for genuine users only (label == 1)
df_genuine = df_features[df_features['label'] == 1].copy()

# 2. Select numeric parametric features only
features_for_clustering = [
    'dwell_gamma_shape', 'dwell_gamma_loc', 'dwell_gamma_scale', 
    'dwell_lognorm_s', 'dwell_lognorm_loc', 'dwell_lognorm_scale',
    'digraph_gamma_shape', 'digraph_gamma_loc', 'digraph_gamma_scale', 
    'digraph_lognorm_s', 'digraph_lognorm_loc', 'digraph_lognorm_scale'
]

X = df_genuine[features_for_clustering].values
user_labels = df_genuine['user_id'].values
trial_labels = df_genuine['trial_id'].values

# 3. Standardize features (Crucial as shape and scale have different units)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Perform Hierarchical Clustering (Ward's Method)
Z = linkage(X_scaled, method='ward')

# 5. Visualize with Dendrogram
plt.figure(figsize=(15, 8))
dendrogram(
    Z,
    labels=[f"{u}_{t}" for u, t in zip(user_labels, trial_labels)],
    leaf_rotation=90,
    leaf_font_size=10,
    color_threshold=0.5 * max(Z[:, 2]) # Optional coloring threshold
)

plt.title('Hierarchical Clustering of User Behavioral Signatures (Genuine Data Only)')
plt.xlabel('User ID _ Trial ID')
plt.ylabel('Ward Distance')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## 6. Evaluating Cluster Purity

To quantitatively and visually assess how well our clusters represent individual users, we cut the dendrogram into $K$ clusters (where $K$ is the number of users) and create a **Purity Heatmap** (Contingency Table).

A "pure" clustering will show a strong diagonal or a single high-value cell per row/column, indicating that each cluster is dominated by a single user's trials.

```python
from scipy.cluster.hierarchy import fcluster

# 1. Extract clusters - we expect USERS_TO_LOAD number of clusters
num_clusters = USERS_TO_LOAD
df_genuine['cluster'] = fcluster(Z, t=num_clusters, criterion='maxclust')

# 2. Create Contingency Table (Crosstab)
contingency_matrix = pd.crosstab(df_genuine['user_id'], df_genuine['cluster'])

# 3. Plot Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(contingency_matrix, annot=True, fmt='d', cmap='YlGnBu', cbar_kws={'label': 'Number of Trials'})

plt.title(f'Cluster Purity Heatmap: Actual Users vs. Predicted Clusters ($K={num_clusters}$)', fontsize=15)
plt.xlabel('Cluster ID (Hierarchical Clustering)', fontsize=12)
plt.ylabel('Actual User ID', fontsize=12)
plt.show()

# 4. Calculate overall Purity Score
purity_score = np.sum(contingency_matrix.max(axis=0)) / len(df_genuine)
print(f"Overall Clustering Purity: {purity_score:.2%}")
```

## 7. Full Dataset Analysis (88 Users)

In this final section, we scale up the analysis to all 88 users in the dataset. We load the pre-extracted features and perform hierarchical clustering to evaluate how well the parametric signatures hold up as the user population grows.

```python
# Load full feature set
FULL_FEATURES_PATH = r'../data_processed/feature_kmt_dataset_Edge_Hill_University_22/feature_extraction_V2/full_parametric_features_88_users.csv'
df_full = pd.read_csv(FULL_FEATURES_PATH)

# Filter for genuine data
df_full_genuine = df_full[df_full['label'] == 1].copy()

print(f"Loaded {len(df_full_genuine)} trials for {len(df_full_genuine['user_id'].unique())} users.")

X_full = df_full_genuine[features_for_clustering].values
user_labels_full = df_full_genuine['user_id'].values

# Scale and Cluster
scaler_full = StandardScaler()
X_scaled_full = scaler_full.fit_transform(X_full)
Z_full = linkage(X_scaled_full, method='ward')

# Plot Dendrogram
plt.figure(figsize=(20, 10))
dendrogram(Z_full, labels=user_labels_full, leaf_rotation=90, leaf_font_size=8)
plt.title('Hierarchical Clustering Dendrogram (Full 88 User Dataset)')
plt.xlabel('User ID')
plt.ylabel('Ward Distance')
plt.show()
```

```python
# Cluster Purity Heatmap
num_clusters_full = len(df_full_genuine['user_id'].unique())
df_full_genuine['cluster'] = fcluster(Z_full, t=num_clusters_full, criterion='maxclust')
contingency_full = pd.crosstab(df_full_genuine['user_id'], df_full_genuine['cluster'])

plt.figure(figsize=(15, 12))
sns.heatmap(contingency_full, annot=False, cmap='YlGnBu')
plt.title('Cluster Purity Heatmap (88 Users)')
plt.xlabel('Cluster ID')
plt.ylabel('Actual User ID')
plt.show()

purity_full = np.sum(contingency_full.max(axis=0)) / len(df_full_genuine)
print(f"Full Dataset Clustering Purity: {purity_full:.2%}")
```

## 8. Advanced Optimization for Behavioral Clustering

To improve the clustering purity (currently ~38% for 88 users), we apply several optimization techniques:
1. **Dimensionality Reduction (PCA & t-SNE)**: To remove noise and visualize high-dimensional clusters.
2. **Alternative Distance Metrics**: Comparing Euclidean with **Cosine Similarity**, which is often more robust for behavioral signatures.
3. **Linkage Optimization**: Evaluating Ward vs. Average/Complete linkage.

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

# 1. PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_full)

plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=user_labels_full, palette='viridis', legend=False, alpha=0.6)
plt.title('PCA Projection of Biometric Features (88 Users)')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.show()

# 2. t-SNE Visualization (Better for Manifold/Non-linear structures)
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled_full)

plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=user_labels_full, palette='tab20', legend=False, alpha=0.8)
plt.title('t-SNE Visualization of User Clustering (88 Users)')
plt.show()
```

```python
# 3. Distance Metric Comparison: Cosine vs. Euclidean

# Try Average Linkage with Cosine Distance
Z_cosine = linkage(X_scaled_full, method='average', metric='cosine')

num_clusters_opt = len(np.unique(user_labels_full))
clusters_cosine = fcluster(Z_cosine, t=num_clusters_opt, criterion='maxclust')

contingency_cosine = pd.crosstab(user_labels_full, clusters_cosine)
purity_cosine = np.sum(contingency_cosine.max(axis=0)) / len(df_full_genuine)

print(f"Purity with Ward/Euclidean: {purity_full:.2%}")
print(f"Purity with Average/Cosine: {purity_cosine:.2%}")

# 4. Cophenetic Correlation Coefficient (Measures how well clustering preserves pairwise distances)
c_euclidean, _ = cophenet(Z_full, pdist(X_scaled_full))
c_cosine, _ = cophenet(Z_cosine, pdist(X_scaled_full, 'cosine'))

print(f"Cophenetic Correlation (Euclidean): {c_euclidean:.3f}")
print(f"Cophenetic Correlation (Cosine): {c_cosine:.3f}")
```

```python
# 5. Threshold Optimization (Elbow Method/Silhouette Analysis Placeholder)
from sklearn.metrics import silhouette_score

thresholds = np.linspace(0.1, 1.5, 10)
scores = []

for t in thresholds:
    labels = fcluster(Z_cosine, t=t, criterion='distance')
    if len(np.unique(labels)) > 1:
        score = silhouette_score(X_scaled_full, labels, metric='cosine')
        scores.append(score)
    else:
        scores.append(np.nan)

plt.figure(figsize=(10, 5))
plt.plot(thresholds, scores, 'bo-')
plt.title('Silhouette Score vs. Clustering Distance Threshold (Cosine)')
plt.xlabel('Distance Threshold')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

## 9. Optimized Cluster Purity Visualization (Cosine Similarity)

Based on the optimizations in Part 8, we visualize the purity of the clusters formed using **Cosine Distance** and **Average Linkage**. This heatmap represents our best effort at unsupervised user separation using parametric behavioral signatures.

```python
# Plot the Optimized Contingency Heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(contingency_cosine, annot=False, cmap='YlGnBu')
plt.title(f'Optimized Cluster Purity Heatmap (Cosine/Average)\nOverall Purity: {purity_cosine:.2%}', fontsize=16)
plt.xlabel('Optimized Cluster ID', fontsize=12)
plt.ylabel('Actual User ID', fontsize=12)
plt.show()

print(f"Final Optimized Purity: {purity_cosine:.2%}")
```

## 10. Final Analysis & Insights

### Why did Ward/Euclidean (38%) slightly outperform Cosine/Average (37%)?

1.  **Discrimination by Magnitude**: In keystroke dynamics, the absolute speed (e.g., a user always having a Dwell Loc of 0.08s vs 0.12s) is a critical identifier. Euclidean distance captures this 'magnitude' difference, while Cosine similarity ignores it in favor of the 'rhythm pattern' only.
2.  **Cluster Compactness**: Ward's method is mathematically optimized to create compact, equal-sized clusters. This aligns well with our dataset structure where every user has a fixed number of trials.
3.  **Feature Nature**: The parametric features extracted (Gamma/Log-Normal) are not sparse. For dense, continuous features like these, Euclidean distance often remains the gold standard.

### Conclusion
The parametric fitting approach successfully provides a biometric 'signature' that is much more sophisticated than simple averages. While unsupervised clustering reaches ~38% purity across 88 users, this provides a very strong baseline for a **Supervised Classifier** (like Random Forest or SVM), which would likely achieve >90% accuracy by learning these specific distribution boundaries.
